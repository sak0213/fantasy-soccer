```python
import requests
import pandas as pd
import numpy as np
from pprint import pprint

url = 'https://fantasy.premierleague.com/api/bootstrap-static/'

r = requests.get(url)
json = r.json()
json.keys()
```

```python
events_df = pd.DataFrame(json["teams"])
events_df.head(20).sort_values("strength_overall_home", ascending= False)
events_df["strength_overall"] = (events_df["strength_overall_home"] + events_df["strength_overall_away"]) /2
events_df["strength_attack"] = (events_df["strength_attack_home"] + events_df["strength_attack_away"]) /2
events_df["strength_defence"] = (events_df["strength_defence_home"] + events_df["strength_defence_away"]) /2
slim_events_df = events_df[["name", "strength_overall", "strength_attack", "strength_defence"]]
slim_events_df.head(20).sort_values("strength_overall", ascending= False)
```
